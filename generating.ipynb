```python
def display_audio(pm: pretty_midi.PrettyMIDI, seconds=30):
  waveform = pm.fluidsynth(fs=16000)
  # Take a sample of the generated waveform to mitigate kernel resets
  waveform_short = waveform[:seconds*16000]
  return display.Audio(waveform_short, rate=16000)
```

```python
import music21 as m21
```

```python
all_notes = [] #loading this data is so aids we'll just stick with a small slice of it, otherwise its just wasting a gpu session
for file in filenames:
  midi = m21.converter.parse(file)
  try:
    s = m21.instrument.partitionByInstrument(midi)
    notes_to_parse = s.parts[0].recurse()
  except:
    notes_to_parse = midi.flat.notes

  for element in notes_to_parse:
    if isinstance(element, m21.note.Note):
      all_notes.append(str(element.pitch))
    elif isinstance(element, m21.chord.Chord):
      all_notes.append('.'.join(str(n) for n in element.normalOrder))
```

```python
n_vocab = len(set(all_notes)) #all the unique midi pitches or chords
n_vocab
```

```python
seq_len = 100 #length of sequence fed into model
pitchnames = sorted(set(all_notes))
note_to_int = dict((note, number) for number, note in enumerate(pitchnames))
network_input = []
network_output = []
```

```python
for i in range(0, len(all_notes) - seq_len, 1):
  sequence_in = all_notes[i:i + seq_len] # creating sequences of 100
  network_input.append([note_to_int[char] for char in sequence_in]) #encoding the sequence with our dict
```

```python
n_patterns = len(network_input)

# reshape the input into a format compatible with LSTM layers
normalized = np.reshape(network_input, (n_patterns, seq_len, 1))
# normalize input
normalized = normalized / float(n_vocab)
```

```python
from tensorflow.keras.models import load_model

model = load_model('/content/drive/MyDrive/checkpoint.model.keras')
```

```python

melody_seq = network_input[0] # value between 0 and size of the network input
generated_notes = []

int_to_note = dict((number, note) for number, note in enumerate(pitchnames)) #reverse of the same array we used for encoding


for i in range(250): #generating 250 notes/chords
  pred_input = np.reshape(melody_seq, (1, len(melody_seq), 1))
  pred_input = pred_input / float(n_vocab)

  pred_note = model.predict(pred_input, verbose=0) # predicting the next note/chord

  index = np.argmax(pred_note) #take value with highest probability
  result = int_to_note[index] #decode it with our dict
  generated_notes.append(result) #append it to our output

  melody_seq = np.append(melody_seq,index) #append it to the sequence so the model moves one step ahead
  melody_seq = melody_seq[1:len(melody_seq)] #ignore the first note to keep a consistent input sequence length
```

```python
offset = 0 # starting "time"
output_notes = []
for sound in generated_notes:
        # current sound is a chord
  if ('.' in sound) or sound.isdigit():
    notes_in_chord = sound.split('.')
    notes = []
    for current_note in notes_in_chord: #splitting the chord into single notes
      new_note = m21.note.Note(int(current_note))
      new_note.storedInstrument = m21.instrument.Piano()
      notes.append(new_note)
    new_chord = m21.chord.Chord(notes)
    new_chord.offset = offset
    output_notes.append(new_chord)
  else: # if its a single note
    new_note = m21.note.Note(sound)
    new_note.offset = offset
    new_note.storedInstrument = m21.instrument.Piano()
    output_notes.append(new_note)

  offset += 0.5 #increasing the offset(otherwise the notes would just be stacked on top of each other)

midi_stream = m21.stream.Stream(output_notes)
midi_stream.write('midi', fp='/content/drive/MyDrive/test_output.mid')
```
